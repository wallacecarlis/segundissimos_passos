# Segundssimos Passos na Anlise de dados - List Comprehension - Aula 2

### Uma **expresso** extremamente til na linguagem **python** que geralmente retorna uma **lista** de dados  a **list comprehension**. Esta tcnica permite realizar diversos tratamentos, inclusive com **condies e verificaes**. Vamos utilizar nossa base de dados de **pilotos ranqueados** para nossos exemplos neste projeto.
### Iniciaremos, por padro, importando as bibliotecas, acessando a base de dados, e providenciaremos cpias para os tratamentos.

1 - Importando as bibliotecas

```python
# 1 - Importando as bibliotecas
import pandas as pd
pd.set_option("mode.chained_assignment", None)
import warnings
warnings.simplefilter("ignore")
```

2 - Acessando a base de dados

```python
# 2 - Acessando e visualizando a base de dados
df = pd.read_csv("https://raw.githubusercontent.com/wallacecarlis/arquivos/main/todos_pilotos_rank.csv")
df
```

3 - Gerando as cpias

```python
# 3 - Criando cpias da base de dados
df_1 = df.copy()
df_2 = df.copy()
```

### Utilizaremos o mtodo **title**, que define cada **primeira letra** de **cada palavra** **no texto**, ou seja, **naquela coluna**, como **maiscula**. No caso, cada **primeira letra** do **nome e sobrenome** de cada piloto ser **maiscula**.
### A sintaxe  bem simples: **x** receber a transformao, para cada **x** em determinada **coluna**.

4 - List Comprehension + title()

```python
# 4 - List Comprehension + title
df_1.nome = [x.title() for x in df_1.nome]
df_1
```

### Faremos o mesmo tratamento com a coluna **equipe**.

```python
# 4.1 - List Comprehension + title()
df_1.equipe = [x.title() for x in df_1.equipe]
df_1
```

### Um segundo modo de alterar letras entre **maisculas e minsculas**  o mtodo **capitalize**. No caso, **somente a primeira letra** ser **maiscula**, todas as demais sero **minsculas**.

5 - List Comprehension + capitalize()

```python
# 5 - List Comprehension + capitalize()
df_1.pais = [x.capitalize() for x in df_1.pais]
df_1
```

### O mtodo **capitalize**  nitidamente percebido quando o texto por si possui palavras que **precisam** de letras **maisculas**.

```python
# 5.1 - List Comprehension + capitalize()
df_1.naturalidade = [x.capitalize() for x in df_1.naturalidade]
df_1
```

### Trataremos a coluna **naturalidade** com o mtodo **title** para verificar tambm nitidamente a diferena para o tratamento com o mtodo **capitalize**.

```python
# 5.2 - List Comprehension + title()
df_1.naturalidade = [x.title() for x in df_1.naturalidade]
df_1
```

### Corrigiremos a coluna naturalidade com a combinao de **title e replace**.

```python
# 5.3 - List Comprehension + capitalize() + replace()
df_1.naturalidade = [x.title().replace("D", "d") for x in df_1.naturalidade]
df_1
```

### Podemos tambm realizar **verificaes** em conjunto com os tratamentos. Para isso, utilizaremos a instruo **if** e seu complemento **else**, caso haja alguma situao se encaixe no tratamento que idealizamos.
### Para exemplificar, iremos procurar o nome **AYRTON SENNA** em letras **maisculas**, e se houver, vamos transform-las em **minsculas** com o mtodo **lower**, caso contrrio, aplicaremos o mtodo **capitalize**.

6 - List Comprehension + lower() + capitalize() + if e else

```python
# 6 - List Comprehension + lower() + capitalize() + if e else
df_2.nome = [x.lower() if x == "AYRTON SENNA" else x.capitalize() for x in df_2.nome]
df_2
```

### Podemos criar uma nova coluna **aplicando** verificaes e condies a partir de outras colunas. Como exemplo, criaremos uma coluna de **salrio** multiplicando a **idade** por **5000** caso a **idade seja par**, ou por **4000** caso seja **mpar**.

7 - List Comprehesion + if e else + ==

```python
# 7 - List Comprehension com if e else
df_2["salario"] = [x * 5000 if x % 2 == 0
                   else x * 4000
                   for x in df_2.idade]
df_2
```

### Criaremos uma nova coluna com o **bnus** sobre o **salrio** com **trs verificaes**: **10%** quando o salrio for **maior que 300 mil**, **5%** quando o salrio for **menor que 150 mil**, e **7%** para o restante dos casos.

8 - List Comprehension + ifs e elses

```python
# 8 - List Comprehension com if, elif e else
df_2["bonus"] = [x * 0.1 if x > 300000
                 else x * 0.05 if x < 150000
                 else x * 0.07
                 for x in df_2.salario]
df_2
```

### Criaremos uma coluna **ranking** verificando **somente o ndice** da base de dados. Utilizaremos para isso a funo **len** que verifica a quantidade total das linhas da base de dados.
### A ttulo de informao, a coluna **rank** apesar de trazer a informao numrica, ainda  um **mtodo**. Por isso a necessidade da criao de uma nova coluna para o **ranking**. Podemos confirmar isso com a funo **type**.

9 - List Comprehension + len()

```python
# 9 - Verificando o len da coluna df_2.rank
len(df_2["rank"])
```

```python
# 9.1 - Verificando os tipos de dados das colunas: idade, rank e salrio
print(type(df_2.idade))
print()
print(type(df_2.rank))
print()
print(type(df_2.salario))
```

```python
# 9.2 - Criando uma coluna para ajustar o ndice do rank (comportamento do row number)
df_2["ranking"] = [len(df_2["rank"]) + 1 - int(i)
                   for i in df_2["rank"]]
df_2
```

### Uma vez que temos de fato uma coluna **ranking**, podemos retornar os nomes para letras **maisculas** e transformar a coluna **bnus** para valores **inteiros**. Faremos isso com o mtodo **upper** e a funo **int**.

10 - List Comprehension + upper() + int()

```python
# 10 - Alterando as colunas nome e bnus com "upper" e "int"
df_2["nome"] = [x.upper() for x in df_2.nome]
df_2["bonus"] = [int(x) for x in df_2.bonus]
df_2
```

### Finalizaremos excluindo a coluna de mtodo **rank**, pois ela j cumpriu seu propsito. Faremos isso com o mtodo **drop**. Em seguida, salvaremos o resultado com o mtodo **to_csv**.

11 - Drop()

```python
# 11 - Excluso da coluna "rank"
df_2.drop("rank", axis = 1, inplace = True)
df_2
```
